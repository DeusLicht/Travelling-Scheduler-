```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from bs4 import BeautifulSoup as Soup
```

```python
map_kml = "Travel.kml"
with open('Travel.kml') as data:
    kml_soup = Soup(data, 'lxml-xml')
coord = kml_soup.find_all('coordinates')
coord_list=[]
for x in coord:
    coord_list.append(str(x))
#coord_list

name = kml_soup.find_all('name')
name_list = []
for x in name:
    name_list.append(str(x))
```

```python
cor_col = ['Longitude', 'Latitude', 'Landmark']
df_col = pd.DataFrame(columns = cor_col)
for i, j in zip(range(len(coord_list)), range(2, len(name_list))):
    itm = coord_list[i]
    string_split = itm.split(',')
    #print(string_split[0][24:])
    cor1 = float(string_split[0][24:])
    cor2 = float(string_split[1])
    cor3 = name_list[j][6:-7]
    df_col.loc[i] = [cor1, cor2, cor3]
```

```python
df_col
```

```python
colormap = plt.cm.rainbow_r
normalize = matplotlib.colors.BoundaryNorm(np.arange(0,15,1), colormap.N)
fig = plt.figure(figsize=(15,6))
plt.scatter(df_col['Longitude'], df_col['Latitude'], cmap = colormap, norm= normalize, s = 150, marker = ".")

plt.xlabel('Latitude', fontsize=14)
plt.ylabel('Longitude', fontsize=14)
plt.title('Scatter plot of Landmarks to Travel', fontsize=14)
plt.show()
```

```python
X = df_col[['Longitude','Latitude']].values
```

```python
#K-Means (Eucliedian Distance)
from sklearn.cluster import KMeans 
km = KMeans(n_clusters=14, random_state=1 )
predictions = km.fit_predict(X)
```

```python
days = pd.concat([df_col.reset_index(), pd.DataFrame({'Cluster':predictions})], axis=1)
#days
days.drop("index", axis = 1, inplace = True)
```

```python
cond_list = [
    days['Cluster'] == 0,
    days['Cluster'] == 1,
    days['Cluster'] == 2,
    days['Cluster'] == 3,
    days['Cluster'] == 4,
    days['Cluster'] == 5,
    days['Cluster'] == 6,
    days['Cluster'] == 7,
    days['Cluster'] == 8,
    days['Cluster'] == 9,
    days['Cluster'] == 10,
    days['Cluster'] == 11,
    days['Cluster'] == 12,
    days['Cluster'] == 13,
]
options = ['Week 1 Sun', 'Week 1 Mon', 'Week 1 Tues', 'Week 1 Wed', 'Week 1 Thurs', 'Week 1 Fri', 'Week 1 Sat', 
           'Week 2 Sun', 'Week 2 Mon', 'Week 2 Tues', 'Week 2 Wed', 'Week 2 Thurs', 'Week 2 Fri', 'Week 2 Sat']
days['Schedule'] = np.select(cond_list, options, default = 'red')
```

```python
days.sort_values(by=['Cluster'])
```

```python
colormap = plt.cm.rainbow
normalize = matplotlib.colors.BoundaryNorm(np.arange(0,14,1), colormap.N)
fig = plt.figure(figsize=(15,6))
plt.scatter(days['Longitude'], days['Latitude'], c = days['Cluster'], cmap = colormap, norm= normalize, s = 150, marker = ".")
plt.colorbar(ticks=np.linspace(0,13,14))
plt.xlabel('Latitude', fontsize=14)
plt.ylabel('Longitude', fontsize=14)
plt.title('k-means clustering (n_clusters=14)', fontsize=14)
plt.show()
```
